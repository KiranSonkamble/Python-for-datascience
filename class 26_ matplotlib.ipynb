### pie chart
 * categorical
*

```python

exp_label=['home rent','food','bill','car','other']
exp_values=[1400,600,400,410,250]

plt.pie(exp_values, labels=exp_label)
```

```python
plt.pie(exp_values, labels=exp_label,radius=1.7)
```

```python

plt.pie(exp_values, labels=exp_label,radius=1.7,autopct='%0.2f%%')
```

```python


plt.pie(exp_values, labels=exp_label,radius=1.7,autopct='%0.1f%%')
```

```python
# '% %'
# '% 0.2%%'
```

```python
# explode rthe chart from the graph
# give equal len of value in explode
plt.pie(exp_values, labels=exp_label,radius=1.7,autopct='%0.2f%%',explode=[0.05,0,0,0,0.11])
```
