### merge  - the data based on some same col

* can merge only 2 df at a time
* at least one col must be same in 2 df


```python
# Ij: common
# lj: complete left
    
# lr: complete right
    

# Oj: complete 2 df
```

```python

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1,df2,how='inner', on='city')
```

```python

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1,df2,on='city')
```

```python
pd.merge(df1,df2,how='left',on='city')
```

```python
pd.merge(df1,df2,on='city',indicator=True)
```
